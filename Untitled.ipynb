```python
#Student Grade Calculator:
def get_positive_int(prompt):
    while True:
        try:
            val = int(input(prompt).strip())
            if val > 0:
                return val
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input. Enter a valid integer.")

def get_mark(subject_index):
    while True:
        try:
            val = float(input(f"Enter marks for subject {subject_index} (0-100): ").strip())
            if 0 <= val <= 100:
                return val
            else:
                print("Marks must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Enter a numeric value.")

def compute_grade(percent):
    if percent >= 90:
        return "A+"
    if percent >= 80:
        return "A"
    if percent >= 70:
        return "B"
    if percent >= 60:
        return "C"
    if percent >= 50:
        return "D"
    if percent >= 40:
        return "E"
    return "F"

def main():
    print("=== Student Grade Calculator ===")
    name = input("Student name: ").strip()
    roll = input("Roll number (optional): ").strip()
    n = get_positive_int("Number of subjects: ")

    marks = []
    for i in range(1, n+1):
        m = get_mark(i)
        marks.append(m)

    total = sum(marks)
    max_total = n * 100
    percentage = (total / max_total) * 100
    grade = compute_grade(percentage)
    status = "Pass" if percentage >= 40 else "Fail"

    # Print report
    print("\n--- Result ---")
    print(f"Name: {name}")
    if roll:
        print(f"Roll: {roll}")
    for i, m in enumerate(marks, start=1):
        print(f"Subject {i}: {m}/100")
    print(f"Total: {total}/{max_total}")
    print(f"Percentage: {percentage:.2f}%")
    print(f"Grade: {grade}")
    print(f"Status: {status}")

    # Optional: save result to file
    save = input("Save result to file? (y/n): ").strip().lower()
    if save == 'y':
        filename = f"{name.replace(' ', '_')}_result.txt"
        with open(filename, "w") as f:
            f.write("=== Student Result ===\n")
            f.write(f"Name: {name}\n")
            if roll:
                f.write(f"Roll: {roll}\n")
            for i, m in enumerate(marks, start=1):
                f.write(f"Subject {i}: {m}/100\n")
            f.write(f"Total: {total}/{max_total}\n")
            f.write(f"Percentage: {percentage:.2f}%\n")
            f.write(f"Grade: {grade}\n")
            f.write(f"Status: {status}\n")
        print(f"Saved to {filename}")

if __name__ == "__main__":
    main()
```
